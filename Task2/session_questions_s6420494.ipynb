```python
import numpy as np
mu,sigma=0.0,1.4
x=np.random.normal(0.0,1.4,100000)
std=np.std(x)
print(std)
import matplotlib.pyplot as plt

count, bins, ignored = plt.hist(x, 30, density=True)

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *

               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),

         linewidth=2, color='y')

plt.show()
```
