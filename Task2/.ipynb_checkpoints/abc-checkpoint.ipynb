```python
from math import *
a=float(input("enter a:"))
b=float(input("enter b:"))
c=float(input("enter c:"))

if a==0:
    if b==0:
        if c==0:
            print(f"infinitely many solutions")
        else:
            print(f"no solutions")
    else:
        x=-c/b
        print(f"solution:{x}")
else:
    D=b**2-4*a*c
    if D<0:
        print(f"no real solutions")
    elif D==0:
        x1=(-1* b + sqrt(D))/(2*a)
        print(f"one solution:{x1}")
    else:
        x1=(-1* b + sqrt(D))/(2*a)
        x2=(-1*b - sqrt(D))/(2*a)
        print(f"two solutions:{x1} and {x2}")
```
