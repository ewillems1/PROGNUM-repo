```python
from scipy.constants import G
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

dictionary={}
for M,r,n in zip(masses,radii,names):
    v_esc=(2*G*M/r)**2
    print(f"escape velocity: {v_esc}")
    print(f" mass: {M:0.2e} ")
    dictionary[n]=v_esc

print(dictionary)
```

```python
from math import *
r=float(input("enter the radius of the cylinder:"))
h=float(input("enter the hight of the cylinder:"))
def cylinder(r,h):
    A=2*pi*(r**2)+2*pi*r*h
    V=pi*(r**2)*h
    return A,V
area,volume=cylinder(r,h)
print(area/volume)
```
